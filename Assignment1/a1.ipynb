## Problem 2


```python
import matplotlib.pyplot as plt
import numpy as np


counts = np.array([200, 100, 300, 50, 150, 200])
sides = np.array([1, 2, 3, 4, 5, 6])

```

```python
# Mass function
pmf = counts / sum(counts)

plt.bar(sides, pmf)
plt.title("Probability Mass Function (PMF)")
plt.xlabel("Side")
plt.ylabel("Probability")
```

```python
cdf = np.cumsum(counts / sum(counts))

plt.step(sides,cdf)
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Side")
plt.ylabel("Probability")
```

## Problem 4.1

```python
D = np.array([[0.1, 1.6, 2.7, 3.2, 4.1, 5.9],
              [0.3, -0.4, -1, -1.6, -2.2, -2.8]]).T

minmax = (D - np.min(D,axis=0)) / (np.max(D) - np.min(D,axis=0))


# print(minmax)
# x,y = minmax
plt.scatter(minmax[:, 0], minmax[:, 1])
plt.title("Min-Max Normalized Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
D_mean = np.mean(D, axis = 0)
D_std = np.std(D, axis=0)
Z_normal = (D - D_mean) / D_std

plt.scatter(Z_normal[:, 0], Z_normal[:, 1])
plt.title("Standard (Z-Score) Normalized Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
