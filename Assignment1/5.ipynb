```python
import matplotlib.pyplot as plt
import numpy as np

SwissRoll = np.loadtxt("swissroll.txt")
Ellipsoids = np.loadtxt("ellipsoids.txt")
Sphere = np.loadtxt("spheres.txt")
# fig = plt.figure()
# ax = plt.axes(projection='3d')
# ax.scatter3D(SwissRoll[:,0], SwissRoll[:,1], SwissRoll[:,2])
# ax.set_xlabel('Feature 1', fontsize=12, fontdict=dict(weight='bold'))
# ax.set_ylabel('Feature 2', fontsize=12, fontdict=dict(weight='bold'))
# ax.set_zlabel('Feature 3', fontsize=12,fontdict=dict(weight='bold'))
# ax.set_title('Swiss Roll', fontsize = 20)


# fig = plt.figure()
# ax = plt.axes(projection='3d')
# ax.scatter3D(Ellipsoids[:,0], Ellipsoids[:,1], Ellipsoids[:,2])
# ax.set_xlabel('Feature 1', fontsize=12, fontdict=dict(weight='bold'))
# ax.set_ylabel('Feature 2', fontsize=12, fontdict=dict(weight='bold'))
# ax.set_zlabel('Feature 3', fontsize=12,fontdict=dict(weight='bold'))

# 1.
SwissRoll_cov_matrix = np.cov(SwissRoll,rowvar=False)
# print(SwissRoll_cov_matrix)

Ellipsoids_cov_matrix = np.cov(Ellipsoids,rowvar=False)
# print(Ellipsoids_cov_matrix)

Sphere_cov_matrix = np.cov(Sphere,rowvar=False)
# print(Sphere_cov_matrix)

# 2.
SwissRoll_eigenvalues, SwissRoll_eigenvectors = np.linalg.eig(SwissRoll_cov_matrix)
# print(SwissRoll_eigenvalues)
# print(SwissRoll_eigenvectors)

Ellipsoids_eigenvalues, Ellipsoids_eigenvectors = np.linalg.eig(Ellipsoids_cov_matrix)
# print(Ellipsoids_eigenvalues)
# print(Ellipsoids_eigenvectors)

Sphere_eigenvalues, Sphere_eigenvectors = np.linalg.eig(Sphere_cov_matrix)
# print(Sphere_eigenvalues)
# print(Sphere_eigenvectors)

```

```python
# 3. SwissRoll
# Perform PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca_result = pca.fit_transform(SwissRoll)

# 2D Projection (first two principal components)
pca_2d = pca_result[:, :2]

# 1D Projection (first principal component)
pca_1d = pca_result[:, 0]

# Plot the 2D projection
# plt.figure(figsize=(8, 6))
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c='blue')
plt.title('2D PCA Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
# plt.grid(True)
plt.show()

# Plot the 1D projection
# plt.figure(figsize=(8, 2))
plt.scatter(pca_1d, np.zeros_like(pca_1d))
plt.title('1D PCA Projection')
plt.xlabel('Principal Component 1')
plt.yticks([])
# plt.grid(True)
plt.show()
```
